# Problem2
In this problem, since the distribution is discreate, we perform sampling for 100 times with different sample number, and calculate the error via the follow formula:
$$error =\frac{1}{N} \sum_{N=100}\sum_{i\in S}(\hat{p_i}-p_i)^2,$$
where $S$ is the support we used for sampling, $\hat{p_i}$ is the estimated probability, $p_i$ is the real probability.

```python
def find(prob,min,max):
    pos_min = prob>min
    pos_max = prob<max
    condition = pos_min & pos_max
    return np.sum(np.where(condition,1,0))

def map2cdf(support,samples,cdf):
    result = []
    result.append(find(samples,0,cdf[0]))
    for i in range(len(support)-2):
        result.append(find(samples,cdf[i],cdf[i+1]))
    result.append(find(samples,cdf[-1],1))
    
    return np.array(result)/len(samples)

def plt_plot(support,cdf,title):
    assert len(support) ==len(cdf)+1
    samples = np.random.rand(SAMPLE_NUM)
    y = map2cdf(support,samples,cdf)
    
    label_y = ["real pmf","1e2","1e3","1e4","1e5"]
    pdf = [cdf[0]]
    for i in range(len(cdf)-1):
        pdf.append(cdf[i+1]-cdf[i])
    pdf.append(1-cdf[-1])
    y_list = []
    y_list.append(pdf)
    
    for j in [100,1000,10000,10000]:
        samples = np.random.rand(j)
        y = map2cdf(support,samples,cdf)
        y_list.append(y)
    
    width = 0; wid = 0.17
    plt.figure(figsize=(12,5))
    clr_list = ['#264653', '#287271', '#2A9D8C', '#8AB07D', '#E9C46B', '#F3A261', '#E66F51']
    for i in range(len(y_list)):  # yj
        plt.bar(np.array(support)+width, y_list[i], width=wid, label=label_y[i],color=clr_list[i])
        width += wid            # 
    plt.legend()

#     plt.bar(support, y, width=0.6, bottom=None,color="deepskyblue")
    plt.title(title)
    plt.show()

    
    
def plt_error_eval(support,cdf,title,repeat_num=100):
    assert len(support) ==len(cdf)+1
    samples = np.random.rand(SAMPLE_NUM)
    y = map2cdf(support,samples,cdf)
    
    label_y = ["1e2","1e3","1e4","1e5"]
    pdf = [cdf[0]]
    for i in range(len(cdf)-1):
        pdf.append(cdf[i+1]-cdf[i])
    pdf.append(1-cdf[-1])
    
    y_list = []
    
    
    for j in [100,1000,10000,10000]:
        err_list = []
        for k in range(repeat_num):
            samples = np.random.rand(j)
            y = map2cdf(support,samples,cdf)
            error = np.sum((y-pdf)**2)
            err_list.append(error)
        y_list.append(np.sum(err_list))
    
    width = 0; wid = 0.17
    plt.figure(figsize=(12,5))
    clr_list = ['#264653', '#287271', '#2A9D8C', '#8AB07D', '#E9C46B', '#F3A261', '#E66F51']
    
    plt.bar(range(len(label_y)), y_list, width=0.5,color=clr_list)
    plt.xticks(range(len(label_y)),labels=label_y)

#     plt.bar(support, y, width=0.6, bottom=None,color="deepskyblue")
    plt.title(title)
    plt.show()
```

## (b) Use Acceptance-Rejection mthod to sample gaussian\
$$ f(x)=\frac{2}{\sqrt{2\pi}}e^{-\frac{x^2}{2}}$$
$$g(x)=e^{-x},c=\sqrt{\frac{2e}{\pi}}$$
$$\frac{f(x)}{cg(x)}=e^{-\frac{(x-1)^2}{2}}$$

```python
def acc_rej(f,g,Y,c):
    sample=[]
    for i in range(len(Y)):
        y = Y[i]
        u = np.random.rand()

        if (f(y)>=c*g(y)*u):
    
            sample.append(y)
    return sample
    
def sing2bil(Y):
    for i in range(len(Y)):
        u = np.random.rand()
        if (u>0.5):
            Y[i] = -Y[i]
    return Y


def abs_normal(x):
    return np.sqrt(2/np.pi)*np.exp(-x**2/2)

def expo(x):
    return np.exp(-x)

def ac_gen(n):
    U = np.random.rand(n)
    c = np.sqrt(2*np.exp(1)/np.pi)
    exp_dist = -np.log(1-U)
    absnormal_dist = acc_rej(abs_normal,expo,exp_dist,c)
    normal = sing2bil(absnormal_dist)
    return normal

f, ax = plt.subplots(2, 2)
f.suptitle('Acceptance-Rejection')
f.subplots_adjust(left=0,right=1,top=0.8,bottom=0,
                    wspace=0.4,hspace=0.3)

sample_e2 = ac_gen(100)
sample_e3 = ac_gen(1000)
sample_e4 = ac_gen(10000)
sample_e5 = ac_gen(100000)

ax[0,0].hist(sample_e2,bins="auto",density=True,range=(-5,5))
ax[0,0].set_title("Sample num = 100")
draw_real_dist(0,0)
ax[0,1].hist(sample_e3,bins="auto",density=True,range=(-5,5))
ax[0,1].set_title("Sample num = 1000")
draw_real_dist(0,1)
ax[1,0].hist(sample_e4,bins="auto",density=True,range=(-5,5))
ax[1,0].set_title("Sample num = 10000")
draw_real_dist(1,0)
ax[1,1].hist(sample_e5,bins="auto",density=True,range=(-5,5))
ax[1,1].set_title("Sample num = 100000")
draw_real_dist(1,1)
plt.show()

# plt.hist(normal,bins="auto",density=True)
# plt.title("Acceptance-Rejection method")
# plt.show()
```

## (b)Use Monte Carlo methods to calculate the intergation $\int_0^1\frac{4}{1+x^2}dx$

```python
from scipy.integrate import quad

def MC_int(f,n,a,b):
    sum=0
    x = np.random.rand(n)*(b-a)+a
    for i in range(len(x)):
        sum+=f(x[i])
    return (b-a)*sum/n

def fx(x):
    return 4/(1+x**2)

theo_ret = quad(fx, 0, 1)
print(theo_ret[0])

print(MC_int(fx,SAMPLE_NUM*100,0,1))

sample_num = [100,200,500,1000,2000,5000,10000,20000,50000,100000]
dis_pool = []
var_pool = []
repeat_time = 20


for num in sample_num:
    dis_list = []
    res_list = []
    for k in range(repeat_time):
        est = MC_int(fx,num,0,1)
        dis_list.append(abs(est-theo_ret[0]))
        res_list.append(est)
    dis_pool.append(np.mean(dis_list))
    var_pool.append(np.var(res_list))
    
plt.plot(sample_num,dis_pool,c="#8AB07D",label="mean")
plt.plot(sample_num,var_pool,c="#E9C46B",label="variance")
plt.xscale("log")
plt.title("error of MC method")
plt.legend()
plt.show()
```
